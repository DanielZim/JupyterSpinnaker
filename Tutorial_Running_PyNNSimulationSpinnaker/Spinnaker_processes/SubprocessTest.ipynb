```python
def isFinished(processes):
    
    for process in processes:
        if(process.poll() == None):
            return False
    
    return True
```

```python
import subprocess
import time
import re

processes = []

# Number of subprocesses
N = 200

# For time measurement
start = time.time()

# Execute all subprocesses and save outputs into files
for i in range(N):
    filename = "output/log" + str(i) + ".txt"
    with open(filename, "w") as file:
        processes.append( subprocess.Popen(['python', 'TestJob.py'], stdout=file) )  

# Wait until all processes are finished
while isFinished(processes) == False:   
    pass

# Time measurement
end = time.time()
print('Time taken in seconds -', end - start)
print("Finished")
```

```python
for i in range(N):

    # Print the content of a file
    with open("output/log" + str(i) + ".txt") as file:
        #print (file.read() )
        filetext = file.read()
    
        # Find fitness value through regex
        matches = re.findall("(.*[^=])=(.*)", filetext)

        #print(type(matches))
        #print(type(matches[0]))
        print(matches[0][1])

print("finished")
```
