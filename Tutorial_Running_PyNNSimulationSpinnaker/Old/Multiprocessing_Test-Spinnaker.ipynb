## Simple Count down:

```python
start = time.time()

for i in range(1):
    countdown()

end = time.time()

print('Time taken in seconds -', end - start)
```

## Same countdown using multiple threads in multiprocessing:

```python
if __name__ == '__main__':
    pool = Pool(processes=100)
    start = time.time()
    
    for i in range(100):
        pool.apply_async(countdown)

    pool.close()
    pool.join()
    
    end = time.time()
    print('Time taken in seconds -', end - start)
```
